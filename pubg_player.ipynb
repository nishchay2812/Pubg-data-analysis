### plot 2 distance + healing -> winning

```python
sns.scatterplot(
    x='Total_Distance',
    y='winPlacePerc',
    hue='Total_healing',
    data=df
)
```


1. Survival is the Best Strategy
The graph shows a much clearer trend than your previous "Kills" plot. It proves that staying mobile is a better predictor of success than combat. The "empty" bottom-left corner confirms that if you don't move, you almost never reach a high win percentage.

2. The "Healing Floor"
Notice the light-colored dots (0–3 heals). They are spread everywhere, including the very bottom of the win percentage. However, the darker dots (9–18 heals) are almost entirely missing from the bottom of the graph.

Insight: High healing acts as a "safety net"—it significantly reduces the chance of a bottom-tier finish.

3. The 6,000-Unit Threshold
There is a visible shift around the 6,000 mark on the "Total_Distance" axis.

Insight: Before 6,000 units, win percentages are scattered randomly. After 6,000 units, the density of dots shifts upward toward the 0.6 to 1.0 range. This suggests 6,000 units is the "break-even" point where a player's chances of a top-tier finish improve drastically.

4. Movement Does Not Guarantee a Win
Even at high distances (10,000+ units), there are still some dots near the bottom of the Y-axis.

Insight: While moving is necessary to win, it isn't a guarantee. This suggests there are "marathon runners" who travel a lot but perhaps die to the blue zone or an ambush before the final circles.
